```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('../data/cities.csv')
lines = pd.read_csv('../data/lines.csv')
station_lines = pd.read_csv('../data/station_lines.csv')
stations = pd.read_csv('../data/stations.csv')
systems = pd.read_csv('../data/systems.csv')
tracks = pd.read_csv('../data/tracks.csv')
track_lines = pd.read_csv('../data/track_lines.csv')
```

```python
stations = stations.dropna(subset = ['name', 'opening', 'closure'])
stations=stations[stations.closure>=9999]
#stations=stations[stations.opening>0]
stations=stations[stations.opening<=2030]
stations.columns=['id','stations_name','geometry','buildstart','opening','closure','city_id']
stations['station_long'] = stations['geometry'].apply(lambda x: x.split('POINT(')[1].split()[1].split(')')[0])
stations['station_lat'] = stations['geometry'].apply(lambda x: x.split('POINT(')[1].split()[0])
```

```python
cities['city_long'] = cities['coords'].apply(lambda x: x.split('POINT(')[1].split()[1].split(')')[0])
cities['city_lat'] = cities['coords'].apply(lambda x: x.split('POINT(')[1].split()[0])
```

```python
city_stations = pd.merge(cities, stations, left_on='id', right_on='city_id', how='inner')
```
